# Creando, Leyendo y Escribiendo datos

En este JNB se revisará el tema de creación, lectura y escritura de datos en Pandas

```python
import pandas as pd
```

# Crear datos

En Pandas los datos que vamos a manejar serán principalmente datos tabulares que puede provenir de diversas fuentes: desde la memoria principal hasta archivos, bases de datos e incluso disponibles a través de las redes.

```python
pd.DataFrame({'Yes': [50, 21], 'No': [131, 2]})
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.', 'Bland.']})
```

```python
pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'],
              'Sue': ['Pretty good.', 'Bland.']},
             index=['Product A', 'Product B'])
```

# Series

A continuación se mostrarán algunos ejemplos de manejo de colecciones de datos mediante el objeto Serie. La Serie se parece a las listas, pero no son exactamente iguales. Coinciden con los Dataframes en que ambos poseen un índice. Aunque -obviamente- no poseen columnas. O bien, puedes considerar que son con 1 columna nada más.

```python
pd.Series([1, 2, 3, 4, 5])
```

```python
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```
